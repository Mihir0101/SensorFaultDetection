## Dropping Unnecessary Columns

```python
cols_needs_to_drop = []

for i in wafer.columns :
    if wafer[i].std() == 0 :
        cols_needs_to_drop.append(i)

cols_needs_to_drop
```

```python
ratios = wafer.isna().sum().div(100)
ratios
```

```python
cols_with_many_nun = list(ratios[ratios > 0.75].index)
cols_with_many_nun
```

```python
wafer.drop(cols_needs_to_drop,axis=1,inplace=True)
wafer.drop(cols_with_many_nun,axis=1,inplace=True)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline = Pipeline(

    steps = [('Imputer',SimpleImputer(strategy='median')),
             ('Scaler',StandardScaler())]
)
```

```python
preprocessor = ColumnTransformer([('NumPipeline',num_pipeline,num_cols)])
```

```python
x = wafer.iloc[:,:-1]
y = wafer.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=53,stratify=y)
```

```python
y_train.value_counts()
```

```python
x_train=pd.DataFrame(preprocessor.fit_transform(x_train),columns=preprocessor.get_feature_names_out())
x_test=pd.DataFrame(preprocessor.transform(x_test),columns=preprocessor.get_feature_names_out())
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=33,k_neighbors=2)

x_train, y_train = sm.fit_resample(x_train,y_train)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
y_train.value_counts()
```

## Model Training

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
models = {

'SVC' : SVC(kernel='rbf', C=1.0, gamma='scale'),
'DT' : DecisionTreeClassifier(criterion='entropy', max_depth=5, class_weight='balanced', random_state=33),
'RF' : RandomForestClassifier(n_estimators=100, 
    random_state=33, 
    class_weight='balanced',
    max_depth=10)
    
}
```

```python
models.items()
```

```python
models.keys()
```

```python
models.values()
```

```python
for i in range(len(models)) :

    model = list(models.values())[i]
    model.fit(x_train,y_train)

    y_predicted = model.predict(x_test)

    reports = classification_report(y_test,y_predicted)

    print(list(models.keys())[i])
    print(reports)
    print("***********************************************")
```

```python
from sklearn.metrics import accuracy_score
```

```python
report = {}

for name,model in models.items() :

    model.fit(x_train,y_train)

    y_predicted = model.predict(x_test)

    accuracy = accuracy_score(y_test,y_predicted)

    report[name] = accuracy
```
