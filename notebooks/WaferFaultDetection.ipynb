## DROPPING UNNECCESSARY COLUMNS

```python
cols_to_drop_std = []

for i in wafer.columns :
    if wafer[i].std() == 0 :
        cols_to_drop_std.append(i)
```

```python
len(cols_to_drop_std)
```

```python
cols_to_drop_std
```

```python
cols_with_ratio = wafer.isna().sum().div(wafer.shape[0])
cols_with_ratio
```

```python
redundant_cols = cols_with_ratio[cols_with_ratio > 0.75].index
redundant_cols = list(redundant_cols)
redundant_cols
```

```python
wafer.drop(cols_to_drop_std,axis=1,inplace=True)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
wafer.head(2)
```

```python
wafer['Good/Bad'].isna().sum()
```

```python
x = wafer.iloc[:,:-1]
y = wafer.iloc[:,-1]
```

## HANDLING MISSING VALUES

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=9)
x = imputer.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=20,random_state=42)
```

```python
sm = SMOTE(random_state=42)
x_train_res,y_train_res = sm.fit_resample(x_train,y_train)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train_res.shape
```

```python
y_train_res.value_counts()
```

## MODEL TRAINING

```python
rf_classifier = RandomForestClassifier()
```

```python
rf_classifier.fit(x_train_res,y_train_res)
```

```python
y_predicted = rf_classifier.predict(x_test)
```

```python
print(classification_report(y_test,y_predicted))
```
